```python
import pandas as pd
import numpy as np           
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/datasets/bank-full.csv",sep=';')
data.shape
```

```python
data.tail()
```

```python
columns_null_values = []
```

```python
def column_null_values(df):
    for column in df:
        number_of_null_values = df[column].notnull().sum()
        print(column," ",number_of_null_values)
        if number_of_null_values < 45211 :
            columns_null_values.append(column)
```

```python
column_null_values(data)
```

```python
def unique_values_column(df):
    for col in df:
        print(f'{col} : {df[col].unique()}')
```

```python
unique_values_column(data)
```

```python

data['marital'].replace({'married' : 0,'single' : 1, 'divorced': 2 },inplace=True)
data['education'].replace({'tertiary' : 0,'secondary' : 1, 'unknown': 2, 'primary': 3 },inplace=True)
data['default'].replace({'no' : 0,'yes' : 1},inplace=True)
data['housing'].replace({'no' : 0,'yes' : 1},inplace=True)
data['loan'].replace({'no' : 0,'yes' : 1},inplace=True)
data['contact'].replace({'unknown' : 0,'cellular' : 1, 'telephone': 2 },inplace=True)

data['poutcome'].replace({'unknown' : 0,'failure' : 1, 'other': 2, 'success': 3 },inplace=True)
data['y'].replace({'no' : 0,'yes' : 1},inplace=True)
data['job'].replace({'management' : 0,'technician' : 1, 'entrepreneur': 2,'blue-collar': 3, 'unknown' : 4,'retired' : 5 ,'admin.': 6,'services' : 7,
                        'self-employed' : 8,  'unemployed' : 9, 'housemaid' : 10,'student' : 11  },inplace=True)
data['month'].replace({'jan' : 0,'feb' : 1, 'mar': 2,'apr': 3, 'may' : 4,'jun' : 5 ,'jul': 6,'aug' : 7,
                        'sep' : 8,  'oct' : 9, 'nov' : 10,'dec' : 11  },inplace=True)
```

```python
unique_values_column(data)
```

```python
scale = ['age','job','marital','education','balance','contact','day','month','duration','campaign','pdays','previous','poutcome']
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler() 
```

```python
data[scale] = scaler.fit_transform(data[scale])
unique_values_column(data)
```

```python
x = data.drop('y',axis='columns')
y = data['y']
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2 , random_state = 1)
```

```python
from sklearn import svm
from sklearn.metrics import classification_report
classifier = svm.SVC(kernel='linear',gamma='auto',C=2)
classifier.fit(x_train,y_train)
y_predict = classifier.predict(x_test);
print(classification_report(y_test,y_predict))

     
```

```python
classifier = svm.SVC(kernel='linear',gamma='auto',C=1)
classifier.fit(x_train,y_train)
y_predict = classifier.predict(x_test);
print(classification_report(y_test,y_predict))
```

```python
classifier = svm.SVC(kernel='linear',gamma='auto',C=100)
classifier.fit(x_train,y_train)
y_predict = classifier.predict(x_test);
print(classification_report(y_test,y_predict))
```

```python
classifier = svm.SVC(kernel='poly',gamma='auto',C=1)
classifier.fit(x_train,y_train)
y_predict = classifier.predict(x_test);
print(classification_report(y_test,y_predict))
```
